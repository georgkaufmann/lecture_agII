## Aufgabe 1.1: Komplexe Zahlen (8P)

- Folgende komplexe Zahlen sind gegeben:
    $$
 \begin{array}{rcl} 
 c_1&=&3-5i \\ c_2&=&-3+4i \\ c_3&=&\sqrt{2}+2i \\ c_4&=&\sqrt{2}-3i
 \end{array}
    $$
- Aufgaben

    a. Zeichnen Sie die vier Zahlen in ein Koordinatensystem (2P).
    
    b. Berechnen Sie: $c_1+c_2$, $c_3-c_1$, $c_3 \cdot c_4$ (2P).
    
    c. Schreiben Sie die allgemeine Formel zur Multiplikation komplexer Zahlen auf (2P).
    
    d. Bilden Sie die komplex konjugierten Zahlen von $c_1$, $c_2$, $c_3$ und $c_4$ (2P).

```python
c1 = 3. - 5j
c2 = -3 + 4j
c3 = np.sqrt(2) + 2j
c4 = np.sqrt(2) - 3j
print(c1,c2,c3,c4)
print(type(c1))
print(c1.real,c1.imag)
```

```python
vectors = np.array([[c1.real,c1.imag],[c2.real,c2.imag],[c3.real,c3.imag],[c4.real,c4.imag]])
print(vectors)
print(vectors.size,vectors.shape)
print(vectors.T[0])
```

```python
vectors = np.array([[c1.real,c1.imag],[c2.real,c2.imag],[c3.real,c3.imag],[c4.real,c4.imag]])
print(vectors)
print(vectors.size,vectors.shape)
print(vectors.T[0])

origin = np.zeros(vectors.size).reshape(vectors.shape)
print(origin)
```

```python
colors = ('blue','green','red','black')
plt.figure(figsize=(6,6))
plt.xlim([-6,6])
plt.ylim([-6,6])

plt.quiver(origin.T[0],origin.T[1],vectors.T[0],vectors.T[1],
           color=colors,angles='xy',scale_units='xy', scale=1)
```

```python
colors = ('blue','green','red','black')
plt.figure(figsize=(6,6))
plt.xlim([-6,6])
plt.ylim([-6,6])

for i in range(vectors.shape[0]):
    plt.arrow(origin.T[0][i],origin.T[1][i],vectors.T[0][i],vectors.T[1][i],width=0.1,color=colors[i])
```
