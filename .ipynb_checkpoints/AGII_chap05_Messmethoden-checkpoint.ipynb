## Mathematisches Pendel

<img src=figures/simples_pendel.jpg style=width:6cm>

Die **Kräftebilanz** für das mathematische Pendel läßt sich einfach ableiten, das nur 
**Gravitationskraft** $F_g$ und **Reibungskraft** $F_r$ wirken:
$$
m \ddot{\vec{r}} = -m \vec{g} - k |v| \vec{v}
$$
Es gilt
$\vec{r}=(x,y,z)$ der Ortsvektor [m],
$\vec{v}=(u,v,w)$ der Geschwindigkeitsvektor [m/s], und
$\vec{g}=(g_x,g_y,g_z)$ der Vektor der Schwerebeschleunigung [m/s$^2$].
$m$ ist die Masse des Pendelgewichts [kg], und $k$ eine Reibungskonstante [kg/m] (Luftreibung).

Betrachten wir zunächst den Fall ohne Reibung ($k=0$), und nutzen für den
Vektor der Schwerebeschleunigung die geometrische Beziehung:
$$
\vec{g} = 
\left(
\begin{array}{c}
\sin(\alpha) g \\ 0 \\ \cos(\alpha) g
\end{array}
\right)
$$ 
mit $\alpha$ [$^{\circ}$] dem Auslenkungswinkel.

- Für $\alpha=0$ (Pendel in Mittelposition) wirkt die Schwerkraft nur in vertikaler Richtung,
$\vec{g}=(0,0,g)$, also senkrecht zur Bewegungsrichtung.
- Für $\alpha \ne 0$ wirkt auf eine Komponente in Bewegungsrichtung:
$\vec{g}=(\sin\alpha g,0,\cos\alpha g)$

Wir interessieren uns nur für die Auslenkung in $x$-Richtung:
$$
m \ddot{x} = -m \sin(\alpha) g
$$
Die Auslenkung in $x$-Richtung läßt sich über den Auslenkungswinkel $\alpha$ beschreiben als
$$
\sin\alpha = {{x}\over{L}}
$$
mit $L$ [m] der Pendellänge.
Für kleine Winkel gilt $\sin\alpha \sim \alpha$, so dass als Bewegungsgleichung folgt:
$$
m \ddot{x} = -m \frac{g}{L} x 
$$
bzw.
$$
 \ddot{x} = - \frac{g}{L} x
$$

Eine analytische Lösung dieser Differentialgleichung zweiter Ordnung ist:
$$
x(t) = A \sin\left( \omega t + \alpha_0 \right)
$$
mit der Amplitude $A$ [m],
der Kreisfrequenz $\omega=\sqrt{{{g}\over{L}}}$,
und der Anfangsauslenkung $\alpha_0$.

```python
# define parameter values
length = 1.0 # m
x0     = 0.1 # m
y0     = length*np.cos(np.arcsin(x0/length))
g      = 9.81
alpha0 = 90.*np.pi/180.
omega  = np.sqrt(g/length)

# define time array
t = np.linspace(0,10,101) # s
alpha  = x0*np.sin(omega*t + alpha0)

# plot
fig,ax = plt.subplots(1,2,figsize=(12,6),sharey=False)
xtop=0.; ytop=0.
xbottom=x0; ybottom=y0
ax[0].set_xlim([-0.3,0.3])
ax[0].set_ylim([1.2,0])
ax[0].set_xlabel('Auslenkung x [m]')
ax[0].set_ylabel('Auslenkung y [m]')
ax[0].plot([xtop,xbottom],[ytop,ybottom],linewidth=5,color='black')
ax[0].plot(xbottom,ybottom,linewidth=0,marker='o',markersize=25,color='red')
#plt.figure(figsize=(12,6))
ax[1].set_xlabel('Zeit t [s]')
ax[1].set_ylabel('Auslenkung A [m]')
ax[1].set_title('Mathematisches Pendel')
ax[1].plot(t,alpha,linewidth=3,color='black',label='analytical')
```

### Berücksichtigung der Reibung 
Differentialgleichung zweiter Ordnung mit Reibungsterm:
$$
m \ddot{x} = -m \frac{g}{L} x - k |v| u
$$
bzw.
$$
 \frac{\partial^2 x}{\partial t^2} = \ddot{x} = - \frac{g}{L} x - \frac{k}{m} |v| u
$$

Diese Differentialgleichung lösen wir numerisch, indem wir die **Differentialgleichung zweiter Ordnung**
zunächst überführen in **zwei Differentialgleichungen erster Ordnung**.
Wir nutzen die Beziehung $\ddot{x}={{\partial^2 x}\over{\partial t^2}}={{\partial u}\over{\partial t}}$:
$$
\begin{array}{rcl}
\frac{\partial u}{\partial t} &=& - \frac{g}{L} x - \frac{k}{m} |v| u \\
\frac{\partial x}{\partial t} &=& u
\end{array}
$$
Nun müssen wir zwei Differentialgleichung erster Ordnung lösen für die beiden **Anfangsbedingungen**
$v_0$ und $x_0$, die Anfangsauslenkung und die Anfangsgeschwindigkeit.

```python
# define parameter values
length    = 1.       # L [m]
friction  = 0.8      # k [kg/m]
gravity   = 9.81     # g [m/s2]
mass      = 0.50     # m [kg]

# define time array
t = np.linspace(0,10,101)

# define initial values
y_initial = [0.,0.1] # u_0 [m/s],x_0 [m]

# define rhs functions
def rhs(y,t,friction,length,mass,gravity):
    # note that the extra argument friction needs to be handed
    # down in a tuple, therefore the extra comma in the call down!
    ynew = []
    vel  = np.sqrt(y[0]**2)
    ynew.append( -gravity/length*y[1] - friction/mass*vel*y[0])
    ynew.append( y[0])
    return ynew

solution1 = scipy.integrate.odeint (rhs,y_initial,t,args=(0.,length,mass,gravity))
solution2 = scipy.integrate.odeint (rhs,y_initial,t,args=(friction,length,mass,gravity))
```

```python
print(solution1.shape)
```

```python
plt.figure(figsize=(12,6))
plt.xlabel('Zeit t [s]')
plt.ylabel('Auslenkung A [m]')
plt.title('Mathematisches Pendel')
plt.plot(t,alpha,linewidth=10,color='lightgrey',label='analytical')
plt.plot(t,solution1[:,1],linewidth=2,color='blue',label='no friction')
plt.plot(t,solution2[:,1],linewidth=2,color='red',label='with friction')
plt.legend(loc='best')
```

Alternative solver `solve_ivp`

```python
def rhs(t,y,friction,length,mass,gravity):
    # note that the extra argument friction needs to be handed
    # down in a tuple, therefore the extra comma in the call down!
    ynew = []
    vel  = np.sqrt(y[0]**2)
    ynew.append( -gravity/length*y[1] - friction/mass*vel*y[0])
    ynew.append( y[0])
    return ynew

solution1 = scipy.integrate.solve_ivp (rhs,[t[0],t[-1]],y_initial,t_eval=t,args=(0.,length,mass,gravity))
solution2 = scipy.integrate.solve_ivp (rhs,[t[0],t[-1]],y_initial,t_eval=t,args=(friction,length,mass,gravity))
#print(solution1.y[0])

plt.figure(figsize=(12,6))
plt.xlabel('Zeit t [s]')
plt.ylabel('Auslenkung A [m]')
plt.title('Mathematisches Pendel')
plt.plot(t,alpha,linewidth=10,color='lightgrey',label='analytical')
plt.plot(t,solution1.y[1],linewidth=2,color='blue',label='no friction')
plt.plot(t,solution2.y[1],linewidth=2,color='red',label='with friction')
plt.legend(loc='best')
```

## Fall

<img src=figures/fall_und_wurf.jpg style=width:6cm>

Die **Kräftebilanz** für den Fall läßt sich einfach ableiten, das nur 
**Gravitationskraft** $F_g$ und **Reibungskraft** $F_r$ wirken:
$$
m \ddot{\vec{r}} = -m \vec{g} - k |v| \vec{v}
$$
Es gilt
$\vec{r}=(x,y,z)$ der Ortsvektor [m],
$\vec{v}=(u,v,w)$ der Geschwindigkeitsvektor [m/s], und
$\vec{g}=(0,0,g)$ der Vektor der Schwerebeschleunigung [m/s$^2$].
$m$ ist die Masse des Fallgewichts, und $k$ eine Reibungskonstante [kg/m] (Luftreibung).

Betrachten wir zunächst den Fall ohne Reibung ($k=0$), dann folgt die
Bewegungsgleichung für Fall in $z$-Richtung:
$$
\ddot z = -g
$$

Analytische Lösung ohne Reibung $k=0$:
$$
\begin{array}{rcl}
v(t) &=& −g t + v_0 \\
z(t) &=& −{{1}\over{2}} g t^2 + v_0 t + z_0
\end{array}
$$

```python
# define parameter values
friction  = 0.8      # k [kg/m]
gravity   = 9.81     # g [m/s2]
mass      = 0.50     # m [kg]

# define time array
t = np.linspace(0,1,101)

# Initial conditions
w_0 = 0
z_0 = 2.

# Analytische Lösung für Geschwindigkeit w und Strecke z
# stoppe Fall bei z=0
w = -gravity*t + w_0
z = -0.5*gravity*t**2 + w_0*t + z_0
for i in range(len(z)):
    if (z[i] < 0):
        z[i]=0.
        w[i]=0.

# plot results
plt.figure(figsize=(12,6))
plt.xlabel('Zeit t [s]')
plt.title('Fall')
plt.grid()
plt.plot(t[w<0],w[w<0],linewidth=2,color='blue',linestyle=':',label='Velocity w [m/s]')
plt.plot(t[z>0],z[z>0],linewidth=2,color='blue',label='Distance z [m]')
plt.legend()
```

### Berücksichtigung der Reibung. 
Bewegungsgleichung für Fall in $z$-Richtung:
$$
\ddot z = -g - \frac{k}{m} |v| w
$$

Zerlege Differentialgleichung zweiter Ordnung in zwei Differentialgleichungen erster Ordnung:
$$
\begin{array}{rcl}
\frac{\partial w}{\partial t} &=& - g - \frac{k}{m} |v| w \\
\frac{\partial z}{\partial t} &=& w
\end{array}
$$
Setze zwei Anfangsbedingungen: $w_0,z_0$.

```python
# define parameter values
friction  = 0.4      # k [kg/m]
gravity   = 9.81     # g [m/s2]
mass      = 0.50     # m [kg]

# define time array
t = np.linspace(0,1,101)

# Initial conditions
y_initial = [0.,2.] # w_0 [m/s],z_0 [m]

# define rhs functions
def rhs(y,t,friction,mass,gravity):
    # note that the extra argument friction needs to be handed
    # down in a tuple, therefore the extra comma in the call down!
    ynew = []
    vel  = np.sqrt(y[0]**2)
    ynew.append( -gravity - friction/mass*vel*y[0])
    ynew.append( y[0])
    return ynew

solution1 = scipy.integrate.odeint (rhs,y_initial,t,args=(0.,mass,gravity))
solution2 = scipy.integrate.odeint (rhs,y_initial,t,args=(friction,mass,gravity))

for i in range(len(solution1)):
    if (solution1[i,1] < 0):
        solution1[i,1]=0.
        solution1[i,0]=0.
for i in range(len(solution2)):
    if (solution2[i,1] < 0):
        solution2[i,1]=0.
        solution2[i,0]=0.
```

```python
# plot results
plt.figure(figsize=(12,6))
plt.xlabel('Zeit t [s]')
plt.title('Fall')
plt.grid()
plt.plot(t,w,linewidth=10,color='lightgray',linestyle=':',label='Analytical w [m/s]')
plt.plot(t,z,linewidth=10,color='lightgray',label='Analytical z [m]')
plt.plot(t,solution1[:,0],linewidth=2,color='blue',label='No friction w [m/s]')
plt.plot(t,solution1[:,1],linewidth=2,color='red',label='No friction z [m]')
plt.plot(t,solution2[:,0],linewidth=2,color='blue',linestyle=':',label='Friction w [m/s]')
plt.plot(t,solution2[:,1],linewidth=2,color='red',linestyle=':',label='Friction z [m]')
plt.legend()
```
