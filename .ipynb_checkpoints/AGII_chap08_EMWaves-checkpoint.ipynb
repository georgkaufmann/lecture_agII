```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.collections import PolyCollection
from matplotlib import ticker, cm
```

## Constants in vaccum

- speed of light $c_0$ [m/s] 
- electrical permittivy $\epsilon_0$ [As/(Vm)]
- magnetic permeability $\mu_0$ [Vs/(Am)]

These quantities are linked by the relation:
$$
c_0 = \frac{1}{\sqrt{\mu_0 \epsilon_0}}
$$

```python
c0       = 2.99792458e8     # m/s
epsilon0 = 8.8541878128e-12 # As/( Vm) 
mu0      = 1./(c0**2 * epsilon0)
print('c0:       ',c0*1e-9,' nm/s')
print('epsilon0: ',epsilon0,' As/Vm')
print('mu0:      ',mu0,' Vs/Am')
```

## Material properties

| material         | $\epsilon_r$     | $\sigma_e$ [mS/m] |
|:----------------:|:----------------:|:---------------------:|
| dry sand         | 4                | $0.01$ |
| wet sand         | 30               | $0.1$  |
| limestone        | 6                | $1$ |

```python
material  = ['dry sand','wet sand','limestone']
epsilon_r = np.array([4,30,6])      # [-]
sigma_e   = np.array([0.01,0.1,1])  # [mS/m]
sigma_e   = sigma_e*1e-3            # [S/m]

print('rel. permittivity:     ',epsilon_r)
print('specific conductivity: ',sigma_e,' S/m')
print('specific resistivity:  ',1/sigma_e,' Omega m')
```

----
## Speed of light in material
$$
c = \frac{c_0}{n}
$$
with $n$ the xx index, defined as
$$
n = \sqrt{\epsilon_r}
$$

----
## Antenna frequency and wavelength

We define a depth section $x$ [m] down to 15m, and use a 100MHz antenna, $f=100$ MHz.
Then,
$$
\omega = 2\pi f
$$


```python
frequency = 100e6        # MHz
omega     = 2*np.pi*frequency
print('frequency:  ',round(frequency/1e6,2),' Hz')

for imaterial in range(len(epsilon_r)):
    n = np.sqrt(epsilon_r[imaterial])
    c = c0/n
    wavelength = c / frequency
    wavenumber = 2*np.pi/wavelength
    print('wavelength: ',round(wavelength,2),'m for material ',material[imaterial])
```

----
## Electro-magnetic wave
$$
E(x,t) = E_0
e^{-\alpha x} 
e^{i \big( \beta x + \omega t \big)} 
$$

```python
x = np.linspace(0,15,401)
t = np.linspace(0,2,21)

def Ez(x,t,wavenumber,omega,A=1):
    Ez = A*np.cos(wavenumber*x - omega*t)
    return Ez

def By(x,t,wavenumber,omega,A=1):
    By = A*np.cos(wavenumber*x - omega*t)
    return By
```

```python
imaterial = 2
n = np.sqrt(epsilon_r[imaterial])
c = c0/n
wavelength = c / frequency
wavenumber = 2*np.pi/wavelength

plt.figure(figsize=(5,10))
plt.ylabel('x [m]')
plt.xlabel('Amplitude')
plt.ylim([15,0])
plt.title('Material: '+str(material[imaterial]))
plt.plot(Ez(x,t[0],wavenumber,omega),x,lw=3,label='no damping')
plt.plot(np.exp(-0.0769*x)*Ez(x,t[0],wavenumber,omega),x,lw=5,label='damping')
plt.grid()
plt.legend()
```

----
## Damping

$$
\alpha =  \omega \sqrt{\frac{\mu \epsilon}{2}  
\left( \sqrt{1+(\frac{\sigma}{\epsilon \omega})^2} - 1 \right)}
$$

```python
def alpha(omega,epsilon_r,sigma_e):
    """
    """
    c0       = 2.99792458e8     # m/s
    epsilon0 = 8.8541878128e-12 # As/( Vm) 
    mu0      = 1./(c0**2 * epsilon0)
    alpha1 = omega * np.sqrt(mu0*epsilon0*epsilon_r/2 * (np.sqrt(1+(sigma_e/epsilon0/epsilon_r/omega)**2) - 1.))
    alpha2 = sigma_e / 2. * np.sqrt(mu0/epsilon0/epsilon_r)
    return alpha1

for imaterial in range(len(epsilon_r)):
    n = np.sqrt(epsilon_r[imaterial])
    c = c0/n
    alpha1 = alpha(omega,epsilon_r[imaterial],sigma_e[imaterial])
    alpha2 = sigma_e[imaterial] / 2. * np.sqrt(mu0/epsilon0/epsilon_r[imaterial])
    print('material: ',material[imaterial],'; c: ',round(c*1e-9,3),' nm/s; alpha: ',round(alpha1,5),round(alpha2,5),' ??')
```

```python
er = np.linspace(1,100,200)
se = np.logspace(-3,5,201)
er,se = np.meshgrid(er,se)
print(er.shape)
a = np.zeros_like(er)

for i in range(er.shape[0]):
    for j in range(er.shape[1]):
        a[i,j] = alpha(omega,er[i,j],se[i,j]*1e-3)
print(a.min(),a.max())

plt.figure(figsize=(8,6))
plt.xlabel('$\sigma_e$ [mS/m]')
plt.ylabel('$\epsilon_r$')
plt.xscale("log")
cbar=plt.contourf(se,er,a,cmap='jet_r', locator=ticker.LogLocator())
marker=['o','s','^']
for i in range(len(epsilon_r)):
    plt.plot(sigma_e[i]*1e3,epsilon_r[i],lw=0,marker=marker[i],markersize=12,markeredgecolor='black',markerfacecolor='grey',alpha=0.5,label=material[i])
plt.colorbar(cbar,label='$\\alpha$ [1/m]')    
plt.legend()
```

```python
fig = plt.figure(figsize=(12,6))
ax = fig.add_subplot(projection='3d')
ax.set_xlabel('x [m]')
ax.set_ylabel('y [m]')
ax.set_zlabel('z [m]')
ax.plot(x,np.exp(-alpha2*x)*Ez(x,t[0]/(frequency),wavenumber,omega),zdir='y',lw=2,label='E(x,t)')
ax.plot(x,np.exp(-alpha2*x)*Ez(x,t[0]/(frequency),wavenumber,omega),zdir='z',lw=2,label='B(x,t)')
ax.legend()
fig.tight_layout()
```

```python
def polygon_under_graph(xlist, ylist):
    """
    Construct the vertex list which defines the polygon filling the space under
    the (xlist, ylist) line graph.  Assumes the xs are in ascending order.
    """
    return [(xlist[0], 0.), *zip(xlist, ylist), (xlist[-1], 0.)]

for i,time in enumerate(t):
    filename = f"EMWave/EMWave1-{i:04}.png"
    print(i,filename)
    EzFilled = []
    ByFilled = []
    EzFilled.append(polygon_under_graph(x, Ez(x,t[i]/(frequency),wavenumber,omega)))
    ByFilled.append(polygon_under_graph(x, By(x,t[i]/(frequency),wavenumber,omega)))

    fig = plt.figure(figsize=(12,6))
    ax = fig.add_subplot(projection='3d')
    ax.view_init(elev=30, azim=45, roll=0)
    ax.set_xlabel('x [m]')
    ax.set_ylabel('y [m]')
    ax.set_zlabel('z [m]')
    poly = PolyCollection(EzFilled, facecolors=['b'], alpha=.6)
    ax.add_collection3d(poly, zs=0, zdir='y')
    ax.plot(x, Ez(x,t[i]/(frequency),wavenumber,omega),zdir='y',color='b',label='E(x,t)')
    poly = PolyCollection(ByFilled, facecolors=['r'], alpha=.6)
    ax.add_collection3d(poly, zs=0, zdir='z')
    ax.plot(x, By(x,t[i]/(frequency),wavenumber,omega),zdir='z',color='r',label='B(x,t)')

    ax.set_xlim(0, 15)
    ax.set_ylim(-1.0,1.0)
    ax.set_zlim(-1.0,1.0)
    ax.set_title('Time: '+str(round(t[i]/frequency*1e9,2))+' ns')
    ax.legend()
    plt.tight_layout()
    plt.savefig(filename,bbox_inches='tight')
    plt.close()
```
