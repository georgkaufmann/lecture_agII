```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([1,2,3,4,5,6])
y = x**2
z = x**3
```

```python
plt.plot(x,y)
```

```python
%matplotlib notebook
fig = plt.figure(figsize=(6,6))
ax = fig.add_subplot(projection='3d')
ax.view_init(elev=20., azim=45.)

ax.set_xlabel('x')
ax.plot(x,y,z)
```
