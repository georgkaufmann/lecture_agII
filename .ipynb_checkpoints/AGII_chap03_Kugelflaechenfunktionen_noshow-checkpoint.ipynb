We first import libraries we will need. This time, besides the standard `numpy` and `matplotlib`
libraries, we need special functions from the `scipy` package.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.integrate
from scipy.special import legendre
from scipy.special import lpn
from scipy.special import lpmn
from scipy.special import sph_harm
from scipy.special import factorial
```

```python
x    = np.linspace(-1,1,51)
nmax = 6
mmax = nmax
```

----
## Legendre polynomials

```python
y = np.zeros(len(x)*(nmax)).reshape(nmax,len(x))
print(y.shape)
for n in range(nmax):
    Pn = legendre(n)
    print(Pn)
    y[n,:] = Pn(x)
```

```python
fig,axs = plt.subplots(2,3,figsize=(12,5))
n=0
axs[0][0].set_xlim([-1.1,1.1])
axs[0][0].set_ylim([-1.1,1.1])
axs[0][0].xaxis.set_tick_params(labelbottom=False)
axs[0][0].yaxis.set_tick_params(labelleft=True)
axs[0][0].plot(x,y[n,:],linewidth=3,color='red',label='P$_{'+str(n)+'}$')
axs[0][0].grid()
axs[0][0].legend()

n=1
axs[0][1].set_xlim([-1.1,1.1])
axs[0][1].set_ylim([-1.1,1.1])
axs[0][1].xaxis.set_tick_params(labelbottom=False)
axs[0][1].yaxis.set_tick_params(labelleft=False)
axs[0][1].plot(x,y[n,:],linewidth=3,color='red',label='P$_{'+str(n)+'}$')
axs[0][1].grid()
axs[0][1].legend()

n=2
axs[0][2].set_xlim([-1.1,1.1])
axs[0][2].set_ylim([-1.1,1.1])
axs[0][2].xaxis.set_tick_params(labelbottom=False)
axs[0][2].yaxis.set_tick_params(labelleft=False)
axs[0][2].plot(x,y[n,:],linewidth=3,color='red',label='P$_{'+str(n)+'}$')
axs[0][2].grid()
axs[0][2].legend()

n=3
axs[1][0].set_xlim([-1.1,1.1])
axs[1][0].set_ylim([-1.1,1.1])
axs[1][0].set_xlabel('x')
axs[1][0].xaxis.set_tick_params(labelbottom=True)
axs[1][0].yaxis.set_tick_params(labelleft=True)
axs[1][0].plot(x,y[n,:],linewidth=3,color='red',label='P$_{'+str(n)+'}$')
axs[1][0].grid()
axs[1][0].legend()

n=4
axs[1][1].set_xlim([-1.1,1.1])
axs[1][1].set_ylim([-1.1,1.1])
axs[1][1].set_xlabel('x')
axs[1][1].xaxis.set_tick_params(labelbottom=True)
axs[1][1].yaxis.set_tick_params(labelleft=False)
axs[1][1].plot(x,y[n,:],linewidth=3,color='red',label='P$_{'+str(n)+'}$')
axs[1][1].grid()
axs[1][1].legend()

n=5
axs[1][2].set_xlim([-1.1,1.1])
axs[1][2].set_ylim([-1.1,1.1])
axs[1][2].set_xlabel('x')
axs[1][2].xaxis.set_tick_params(labelbottom=True)
axs[1][2].yaxis.set_tick_params(labelleft=False)
axs[1][2].plot(x,y[n,:],linewidth=3,color='red',label='P$_{'+str(n)+'}$')
axs[1][2].grid()
axs[1][2].legend()

plt.tight_layout()
plt.savefig('images/pn.png',dpi=400,bbox_inches='tight')
```

----
## Associated Legendre polynomials

```python
y = np.zeros(len(x)*nmax*mmax).reshape(nmax,mmax,len(x))
print(y.shape)
for n in range(nmax):
    for m in range(n+1):
        for i in range(len(x)):
            Pnm = lpmn(m,n,x[i])
            if (m == 0):
                norm = np.sqrt(2*(2*n+1))
            else:
                norm = np.sqrt((2*n+1)*factorial(n-m)/factorial(n+m))
            y[n,m,i] = norm*Pnm[0][m,n]
```

```python
fig,axs = plt.subplots(2,3,figsize=(12,5))
n=5;m=0
axs[0][0].set_xlim([-1.1,1.1])
axs[0][0].set_ylim([-4.1,4.1])
axs[0][0].xaxis.set_tick_params(labelbottom=False)
axs[0][0].yaxis.set_tick_params(labelleft=True)
axs[0][0].plot(x,y[n,m,:],linewidth=3,color='red',label='P$_{'+str(n)+str(m)+'}$')
axs[0][0].grid()
axs[0][0].legend()

n=5;m=1
axs[0][1].set_xlim([-1.1,1.1])
axs[0][1].set_ylim([-4.1,4.1])
axs[0][1].xaxis.set_tick_params(labelbottom=False)
axs[0][1].yaxis.set_tick_params(labelleft=False)
axs[0][1].plot(x,y[n,m,:],linewidth=3,color='red',label='P$_{'+str(n)+str(m)+'}$')
axs[0][1].grid()
axs[0][1].legend()

n=5;m=2
axs[0][2].set_xlim([-1.1,1.1])
axs[0][2].set_ylim([-4.1,4.1])
axs[0][2].xaxis.set_tick_params(labelbottom=False)
axs[0][2].yaxis.set_tick_params(labelleft=False)
axs[0][2].plot(x,y[n,m,:],linewidth=3,color='red',label='P$_{'+str(n)+str(m)+'}$')
axs[0][2].grid()
axs[0][2].legend()

n=5;m=3
axs[1][0].set_xlim([-1.1,1.1])
axs[1][0].set_ylim([-4.1,4.1])
axs[1][0].set_xlabel('x')
axs[1][0].xaxis.set_tick_params(labelbottom=True)
axs[1][0].yaxis.set_tick_params(labelleft=True)
axs[1][0].plot(x,y[n,m,:],linewidth=3,color='red',label='P$_{'+str(n)+str(m)+'}$')
axs[1][0].grid()
axs[1][0].legend()

n=5;m=4
axs[1][1].set_xlim([-1.1,1.1])
axs[1][1].set_ylim([-4.1,4.1])
axs[1][1].set_xlabel('x')
axs[1][1].xaxis.set_tick_params(labelbottom=True)
axs[1][1].yaxis.set_tick_params(labelleft=False)
axs[1][1].plot(x,y[n,m,:],linewidth=3,color='red',label='P$_{'+str(n)+str(m)+'}$')
axs[1][1].grid()
axs[1][1].legend()

n=5;m=5
axs[1][2].set_xlim([-1.1,1.1])
axs[1][2].set_ylim([-4.1,4.1])
axs[1][2].set_xlabel('x')
axs[1][2].xaxis.set_tick_params(labelbottom=True)
axs[1][2].yaxis.set_tick_params(labelleft=False)
axs[1][2].plot(x,y[n,m,:],linewidth=3,color='red',label='P$_{'+str(n)+str(m)+'}$')
axs[1][2].grid()
axs[1][2].legend()

plt.tight_layout()
plt.savefig('images/pnm.png',dpi=400,bbox_inches='tight')
```
