```python
import numpy as np
import matplotlib.pyplot as plt
```

Define electrical conductivities of **fluid** and **matrix**, $\sigma_F$ and $\sigma_M$ [S/m],
and porosity $\Phi$.

```python
sigmaF = 1000.
sigmaM = 1.
porosity = np.logspace(-3,0,101)
```

----
## Archie

The **effective electrical conductivity** $\sigma_{eff}$ [S/m] of a two-component rock (e.g. sediment with pore space)
can be estimated from **Archie's law**:
$$
\sigma_{eff} = \sigma_M + a \Phi^n S^m \sigma_F
$$
with
- $\sigma_M$ [S/m] electrical conductivity of matrix
- $\sigma_F$ [S/m] electrical conductivity of fluid
- $\Phi$ [-] porosity
- $S$ [-] saturation
- $a$, $n$, $m$ empirical constants.

```python
def sigmaArchie(sigmaF,sigmaM,Phi,S=1.,a=1,n=1.5,m=1):
    sigmaArchie = sigmaM + a*sigmaF * Phi**n * S**m
    return sigmaArchie
```

```python
plt.xlabel('Porosity $\Phi$')
plt.ylabel('$\sigma_{eff}$')
plt.xscale("log")
plt.yscale("log")
plt.title('Archie law')
plt.plot(porosity,sigmaArchie(sigmaF,sigmaM,porosity),lw=4,ls='solid',color='black',label='Archie (a=1,S=1,n=1.5,m=1)')
plt.plot(porosity,sigmaArchie(sigmaF,sigmaM,porosity,n=1),lw=2,ls='dashed',color='black',label='Archie (a=1,S=1,n=1,m=1)')
plt.plot(porosity,sigmaArchie(sigmaF,sigmaM,porosity,n=2),lw=2,ls='dotted',color='black',label='Archie (a=1,S=1,n=2,m=1)')

plt.plot(porosity,sigmaArchie(sigmaF,sigmaM,porosity,a=0.5),lw=2,ls='dashed',color='red',label='Archie (a=0.5,S=1,n=1.5,m=1)')
plt.plot(porosity,sigmaArchie(sigmaF,sigmaM,porosity,a=1.5),lw=2,ls='dotted',color='red',label='Archie (a=1.5,S=1,n=1.5,m=1)')
plt.grid()
plt.legend()
plt.savefig('images/ArchieLaw.png',bbox_inches='tight')
```

## Hashin-Shtrikman

In the **Hashin-Shtrikman law**,
the **effective electrical conductivity** $\sigma_{eff}$ [S/m] is bounded by a minimum and a maximum value:
$$
\begin{array}{rcl}
\sigma_{min} &=& \sigma_M \big( 1 + \frac{3 \Phi (\sigma_f - \sigma_M)}{3 \sigma_M + (1-\Phi) (\sigma_f-\sigma_M)} \big) \\
\sigma_{max} &=& \sigma_F \big( 1 - \frac{3 (1-\Phi) (\sigma_f - \sigma_M)}{3 \sigma_F - \Phi (\sigma_f-\sigma_M)} \big)
\end{array}
$$
with
$$
\sigma_{min} \le \sigma_{eff} \le \sigma_{max}
$$

```python
sigmaMin = sigmaM*(1 + (3*porosity*(sigmaF-sigmaM)) / (3*sigmaM + (1-porosity)*(sigmaF-sigmaM)) )
sigmaMax = sigmaF*(1 - (3*(1-porosity)*(sigmaF-sigmaM)) / (3*sigmaF - porosity*(sigmaF-sigmaM)) )

sigmaMin2 = (porosity/sigmaF + (1-porosity)/sigmaM)**(-1)
sigmaMax2 = porosity*sigmaF + (1-porosity)*sigmaM
```

```python
plt.xlabel('Porosity $\Phi$')
plt.ylabel('$\sigma_{eff}$')
plt.xscale("log")
plt.yscale("log")
plt.title('Hashin-Shtrikman law')
plt.plot(porosity,sigmaMin,lw=4,color='red',label='$\sigma_{min}$')
plt.plot(porosity,sigmaMax,lw=4,color='blue',label='$\sigma_{max}$')
plt.fill_between(porosity,sigmaMax,sigmaMin,color='grey',alpha=0.3)
plt.plot(porosity,sigmaMin2,lw=1,ls='dashed',color='black')
plt.plot(porosity,sigmaMax2,lw=1,ls='dashed',color='black')
plt.plot(porosity,sigmaArchie(sigmaF,sigmaM,porosity),lw=2,ls='solid',color='black',alpha=0.5,label='Archie')
plt.grid()
plt.legend()
plt.savefig('images/HashinShtrikmanLaw.png',bbox_inches='tight')
```
