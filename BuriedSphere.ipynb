```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def boug_sphere(x,D=100.,R=50.,drho=500.):
    # Bouguer gravity of solid sphere
    G    = 6.672e-11      # m^3/kg/s^2
    boug = 4./3.*np.pi*G*drho * R**3*D/(x**2+D**2)**(3/2)
    return boug
```

```python
radii = np.linspace(20,150,20)
boug200 = np.zeros_like(radii)
boug300 = np.zeros_like(radii)

for i,R in enumerate(radii):
    boug200[i] = boug_sphere(0.,D=200.,R=R,drho=-1200.)
    boug300[i] = boug_sphere(0.,D=300.,R=R,drho=-1200.) 
```

```python
plt.plot(radii,1.e5*boug200,label='D=200m')
plt.plot(radii,1.e5*boug300,label='D=300m')
plt.xlabel('Radius [m]')
plt.ylabel('Bouguer gravity [mGal]')
plt.legend()
plt.grid()
```
